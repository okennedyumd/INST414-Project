```python
resp2 = requests.get("https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/seasons/2021/types/2/leaders?limit=50")
resp2.status_code
```

```python
resp2_obj = resp2.json()
resp2_obj
```

```python
players2 = (resp2_obj['categories'][2]['leaders'])
players3 = pd.json_normalize(players2)
players4 = players3[['value', 'athlete.$ref', 'team.$ref']].dropna()
players4.head()
```

```python
i = 0
lead = players4['athlete.$ref']
ath1 = []
ath2 = []
for line in lead: 
    tb = requests.get(lead[i])
    tb2 = tb.json()
    a = tb2['id']
    b = tb2['fullName']
    ath1.append(a)
    ath2.append(b)
    i+=1 
```

```python
i = 0
teamname = players4['team.$ref']
ath3 = []
ath4 = []
for line in teamname: 
    tb3 = requests.get(teamname[i])
    tb4 = tb3.json()
    c = tb4['id']
    d = tb4['displayName']
    ath3.append(c)
    ath4.append(d)
    i+=1
```

```python
athelete=pd.DataFrame()   
athelete['PlayerID'] = ath1
athelete['PlayerName'] = ath2
athelete['TeamId'] = ath3
athelete['TeamName'] = ath4
athelete['PassingYards'] = players4['value']

athelete
```

```python
athelete.to_csv('receivingyard.csv')
```

```python
resp4 = requests.get('https://sports.core.api.espn.com/v2/sports/football/leagues/nfl/seasons/2022/draft/rounds')
resp4.status_code
```

```python
resp4_obj = resp4.json()
resp4_obj
```

```python
p = (resp4_obj['items'][6]['picks'])
pl = pd.json_normalize(p)
pla = pl[['athlete.$ref']].dropna()
pla.head()
```

```python
i = 0
lead = pla['athlete.$ref']
ath1 = []
ath2 = []
ath3 = []
for line in lead: 
    tb = requests.get(lead[i])
    tb2 = tb.json()
    a = tb2['id']
    b = tb2['fullName']
    c = tb2['position']['abbreviation']
    ath1.append(a)
    ath2.append(b)
    ath3.append(c)
    i+=1 
```

```python
athelete1=pd.DataFrame()   
athelete1['PlayerID'] = ath1
athelete1['PlayerName'] = ath2
athelete1['Position'] = ath3
```

```python
draft2021 = pd.DataFrame()
```

```python
draft2021 = pd.concat([draft2021, athelete1])
```
